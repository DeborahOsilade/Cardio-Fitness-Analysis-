```python
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
import matplotlib as plt
```

```python
cardio = pd.read_csv("CardioGoodFitness.csv")
```

```python
cardio[cardio['Gender']=='Male']['Product'].value_counts()
```

```python
cardio.describe()
# Usage & Fitness - The mean and median are slightly close 
# There are extreme values in Miles, Income and Age 
```

```python
sns.distplot(cardio['Income'])
#skewed to the right, shows presence of extreme values,50% of customers earn above 50 thousand, 
# and most of the customers earn above $50,000
```

```python
sns.jointplot(cardio['Income'], cardio['Miles'])
```

```python
sns.jointplot(cardio['Miles'], cardio['Usage'])
#customers who uses a product up to 7 times goes up to 300 miles 
# customers mostly use the product 4 times
#customers who use the product twice a week can go up to 100 miles
#most of the customers use the products between 50 and 100 miles, only extreme customers go above 150 miles 
```

```python
sns.jointplot(cardio['Usage'],cardio['Product'])
#product TM798 is the highest used product by Men and Women at the gym  
```

```python
sns.pairplot(cardio[['Usage','Miles','Age', 'Fitness','Product']])
```
